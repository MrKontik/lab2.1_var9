```python
def simuv(data,u,v):
    u=data[u-1:u].values[0][1:]
    u[u<0]=0
    v=data[v-1:v].values[0][1:]
    v[v<0]=0
    return u.dot(v)/math.sqrt(u.dot(u)*v.dot(v))
```

```python
def ru(data,u):
    u=data[u-1:u].values[0][1:]
    u[u<0]=0
    return np.mean(u)
```

```python
def kNN(data,user,k):
    res=np.array([[0,0]])
    u=data[user-1:user].values[0][1:]
    res=np.append(res, [[0,simuv(data,user+1,1)]], axis = 0)
    res=res[1:]
    for i in range(1,40):
        if (i!=user-1):
            res=np.append(res, [[i,simuv(data,user,i+1)]], axis = 0)
        else:
            res=np.append(res, [[i,0]], axis = 0)
    arr=res
    arr=arr[arr[:, 1].argsort()]
    arr=arr[40-k:]
    return arr
```

```python
def rui(data,user,k,i):
    sum_sim=0
    sum_abs_sim=0
    u=data[user-1:user].values[0][1:]
    u[u<0]=0
    knn=kNN(data,user,k)
    friend=np.array([knn[0][0],knn[1][0],knn[2][0],knn[3][0]])
    friend=friend.astype(int)
    for g in range(1,k+1):
        v=data[friend[g-1]:friend[g-1]+1].values[0][1:]
        v[v<0]=0
        sum_sim=sum_sim+simuv(data,user,g+1)*(v[i+1]-ru(data,g+1))
        sum_abs_sim=sum_abs_sim+np.absolute(simuv(data,user,g+1))
    return ru(data,user)+sum_sim/sum_abs_sim
```
